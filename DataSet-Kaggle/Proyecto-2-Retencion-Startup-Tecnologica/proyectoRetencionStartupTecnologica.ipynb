```python


def contar_ceros_y_graficar(df_mensuall, columnas_a_convertir, yPrimaria):
    resultados = {}

    for columna in columnas_a_convertir:
        count = df_mensuall[(df_mensuall[columna] == 0) & (df_mensuall['Retiro_Churn'] == 0)].shape[0]
        resultados[columna] = count if count > 0 else None



    # Convertir datos a listas ordenadas por mes
    meses = list(yPrimaria.keys())  # Eje X
    valores_yPrimaria = list(yPrimaria.values())  # Facturación (Verde)

    # Convertir los resultados de contar_ceros a listas
    valores_ySecundaria = [resultados.get(mes, 0) for mes in meses]  # Valores de la secundaria

    # Crear figura
    plt.figure(figsize=(10, 5))
    plt.plot(meses, valores_yPrimaria, color="green", marker="o", linestyle="-", label="Facturación")
    plt.plot(meses, valores_ySecundaria, color="blue", marker="s", linestyle="-", label="Retirados")

    # Agregar etiquetas con los valores
    for i, txt in enumerate(valores_yPrimaria):
        plt.text(meses[i], valores_yPrimaria[i], str(txt), fontsize=10, ha="left", va="bottom", color="green")

    for i, txt in enumerate(valores_ySecundaria):
        plt.text(meses[i], valores_ySecundaria[i], str(txt), fontsize=10, ha="left", va="top", color="blue")

    # Configuración de etiquetas y título
    plt.xlabel("Meses (2019-2020)")
    plt.ylabel("Valores")
    plt.title("Comparación de Facturación y Retirados")
    plt.legend(loc="upper left")

    plt.xticks(rotation=45)  # Rotar etiquetas del eje X para mejor lectura
    plt.grid(True, linestyle="--", alpha=0.5)

    plt.show()

    return resultados

# Lista de columnas a analizar
columnas_a_convertir = ['1/2019', '2/2019', '3/2019', '4/2019', '5/2019', '6/2019', '7/2019',
                         '8/2019', '9/2019', '10/2019', '11/2019', '12/2019', '1/2020', '2/2020',
                         '3/2020', '4/2020', '5/2020', '6/2020', '7/2020', '8/2020', '9/2020',
                         '10/2020', '11/2020', '12/2020']

# Llamar a la función con df_mensuall y yPrimaria
contar_ceros_y_graficar(df_mensuall, columnas_a_convertir, yPrimaria)
```

## ¡Perfecto! Como tienes más de 1,000 datos en total, usaremos la división 70/30.

Paso 1: Separar X e y
Ahora define:
✅ X = Todas las columnas relevantes para predecir el churn (sin incluir la columna Retiro_Churn).
✅ y = La columna objetivo Retiro_Churn (1 = se retiró, 0 = sigue activo).

```python
df_mensuall_entranamiento = df_mensuall.drop(columns=['Mes de Abandono (Churn)'])  # Creamos un nuevo DataFrame sin esta columna
columnas_binarias = [
    '1/2019', '2/2019', '3/2019', '4/2019', '5/2019', '6/2019', '7/2019',
    '8/2019', '9/2019', '10/2019', '11/2019', '12/2019', '1/2020', '2/2020',
    '3/2020', '4/2020', '5/2020', '6/2020', '7/2020', '8/2020', '9/2020',
    '10/2020', '11/2020', '12/2020'
]
X = df_mensuall[columnas_binarias]
y = df_mensuall_entranamiento['Retiro_Churn']  # Variable objetivo
```

📌 Explicación rápida:

1. X = Todas las columnas excepto Retiro_Churn (variables independientes).
2. y = La columna Retiro_Churn (variable objetivo, 0 o 1).
3. train_test_split divide los datos en 70% entrenamiento y 30% prueba:
4. random_state=42 → Para asegurar que siempre se obtenga la misma división.
5. stratify=y → Para que ambas clases (0 y 1) se mantengan proporcionalmente en ambos conjuntos.

## 1️⃣ Dividir los datos en Train y Test
Usamos train_test_split() para dividir los datos en entrenamiento (70%) y prueba (30%):

```python
from sklearn.model_selection import train_test_split
```

## DEFINICION DE VARIBAÑES

```python
# División de datos: 70% entrenamiento, 30% prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

📌 Explicación:

test_size=0.3 → 30% de los datos para prueba, 70% para entrenamiento.
random_state=42 → Para que la división sea reproducible.
stratify=y → Para mantener la proporción de clases en la división (importante si los datos están desbalanceados).

## 3️⃣ Elegir y entrenar un modelo de Machine Learning 🤖
Dependiendo del análisis, puedes probar diferentes modelos. Algunos comunes para problemas de clasificación como Churn son:

1. ✅ Regresión Logística: Simple, útil para interpretar los resultados.
2. ✅ Árboles de Decisión: Fácil de visualizar y entender.
3. ✅ Random Forest: Mejor precisión y robustez.
4. ✅ XGBoost: Muy eficiente en datos grandes.

## SELECCION DE MODELO
3. ✅ Random Forest: Mejor precisión y robustez.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Crear y entrenar el modelo Random Forest
modelo_rf = RandomForestClassifier(n_estimators=100, random_state=42)
modelo_rf.fit(X_train, y_train)

# Hacer predicciones
y_pred = modelo_rf.predict(X_test)

# Evaluar el modelo
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:\n", classification_report(y_test, y_pred))
```

📌 ¿Qué significan estos valores?
1. 🔹 Accuracy (Precisión General): → 0.9899 (98.99%)
2. 🔹 Precision (Precisión por Clase):

3. Clase 0 (No Churn) → 100% de precisión.
4. Clase 1 (Churn) → 99% de precisión.
5. 🔹 Recall (Sensibilidad / Detección Correcta):
6. Clase 0 → 97% (3% de falsos negativos).
7. Clase 1 → 100% (se detectaron todos los casos correctamente).
8. 🔹 F1-score (Balance entre Precision y Recall):
9. Clase 0 → 98%
10. Clase 1 → 99%

## EVAULACION DE MODELO GAFRICAMENTE

¡Genial! Vamos a generar los siguientes gráficos para evaluar el rendimiento del modelo:

1. 1️⃣ Matriz de Confusión → Para ver cómo clasifica los casos correctamente y dónde se equivoca.
2. 2️⃣ Curva ROC y AUC → Para evaluar la capacidad del modelo en separar clases.
3. 3️⃣ Importancia de Variables → Para ver qué meses afectan más la predicción.

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, roc_curve, auc
import numpy as np
```

```python
# 1️⃣ Matriz de Confusión
y_pred = modelo_rf.predict(X_test)  # Suponiendo que 'model' es tu modelo entrenado
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['No Churn', 'Churn'], yticklabels=['No Churn', 'Churn'])
plt.xlabel('Predicción')
plt.ylabel('Real')
plt.title('Matriz de Confusión')
plt.show()

# 2️⃣ Curva ROC y AUC
y_prob = modelo_rf.predict_proba(X_test)[:, 1]  # Probabilidades de clase positiva
fpr, tpr, _ = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)
plt.figure()
plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Curva ROC')
plt.legend(loc='lower right')
plt.show()

# 3️⃣ Importancia de Variables (Feature Importance)
feature_importance = modelo_rf.feature_importances_  # Suponiendo que es un modelo basado en árboles
features = X.columns
sorted_idx = np.argsort(feature_importance)
plt.figure(figsize=(10, 6))
plt.barh(range(len(sorted_idx)), feature_importance[sorted_idx], align='center')
plt.yticks(range(len(sorted_idx)), np.array(features)[sorted_idx])
plt.xlabel('Importancia')
plt.title('Importancia de Variables')
plt.show()
```
