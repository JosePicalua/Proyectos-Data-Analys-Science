```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

path_csv = r"C:\Users\josep\OneDrive\Escritorio\Proyectos-Junior\Dash\Proyecto-Automobile\PrimeraSolicitud\csv\PrimeraSolicitud.csv"

# Cargar el archivo CSV
df = pd.read_csv(path_csv)

# Filtrar solo los registros de BMW
df_individual = df[df['make'] == 'dodge']

# Crear la gráfica de barras
plt.figure(figsize=(10, 5))
plt.bar(df_individual.index, df_individual['price'], color='blue')

# Etiquetas y título
plt.xlabel(df_individual['make'])
plt.ylabel("Precio")
plt.title("Precios de vehículos BMW")
plt.xticks(rotation=90)

# Mostrar la gráfica
plt.show()
```
